# SETUP PATHS HERE:

```python
import os
import glob

parent_dir = "/content/drive/MyDrive/PATHTO/DATASET" # @param {type:'string'}
output_file = "/content/drive/MyDrive/PATHTO/DATASET/Captions.txt" # @param {type:'string'}
```

# Run Caption Concatenator

```python
# Open the output file in write mode with utf-8 encoding
with open(output_file, "w", encoding="utf-8") as outfile:
    # Recursively traverse the parent directory
    for dirpath, dirnames, filenames in os.walk(parent_dir):
        for filename in filenames:
            # Check if the file is a text file
            if filename.endswith(".txt"):
                # Open the text file with utf-8 encoding
                with open(os.path.join(dirpath, filename), "r", encoding="utf-8") as infile:
                    # Read the contents of the text file and write them to the output file
                    for line in infile:
                        outfile.write(line)
                    outfile.write("\n") # add a newline character after each file
                    print(f"Finished processing {os.path.join(dirpath, filename)}")
            # Ignore .png files
            elif filename.endswith(".png"):
                continue
            # Print a message for other file types
            else:
                print(f"Ignoring file {filename}")
    print("Finished processing all folders")
```
